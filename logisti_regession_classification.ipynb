```python
import numpy as np
import  pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ds=pd.read_csv("Social_Network_Ads.csv")
ds
```

```python
ds.columns
```

```python
#calculate
ds.duplicated().sum()
```

```python
#removed
ds.drop_duplicates(inplace=True)
```

```python
ds.isnull().sum()
```

```python
x=ds[['Age','EstimatedSalary']]
y=ds[['Purchased']]
```

```python
x
```

```python
y
```

```python
#feature
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=pd.DataFrame(sc.fit_transform(x),columns=['Age','EstimatedSalary'])
```

```python
#split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=0)
```

```python
len(x_test)
```

```python
#applying

from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
#predict

myinput=pd.DataFrame(sc.transform([[30,150000]]),columns=['Age','EstimatedSalary'])
classifier.predict(myinput)
```

```python
from sklearn.metrics import accuracy_score
y_pred=classifier.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(f"Accuracy of the logistics regression model:{accuracy * 100:.2f}%")
```

```python
#User input

age=float(input("Enter Age"))
salary=float(input("Enter the salary:"))
myinput=pd.DataFrame(sc.transform([[age,salary]]),columns=['Age','EstimatedSalary'])
result=classifier.predict(myinput)
if result[0]==1:
    print("prediction:user will purchase plan")
else:
    print("prediction:user will not purchase plane")
    
```
